Envía el archivo *names_2009-08-30.txt* como input al comando *cat*, y el resultado de eso (que es la información de names_2009-08-30.txt) lo envía como output al archivo *names_cat*. Como en este caso no existía simplemente lo creó.

Hubiera pasado lo mismo escribiendo solamente *cat names_2009-08-30.txt > names_cat*.

```bash
cat < names_2009-08-30.txt >> names_cat.txt
```

```bash
cat names_cat.txt
```

Hace lo mismo que el anterior, solo que *>>* es para concatenar la información de *names_2009-08-30* al final de *names_cat.txt*, sin sobreescribir lo que ya existía. En este caso simplemente duplicamos la información.

La diferencia entre > y >> es que el primero sobreescribe la información que ya existía y el segundo no.

El comando **umask** se usa para determinar la configuración default de permisos al crear archivos nuevos. Puede especificarse tanto de forma simbólica como numérica. Si se especifican en formato octal debe ponerse un *0* al inicio para especificar que los números siguientes se refieren a permisos.

Por ejemplo:

```bash
umask 0664
```

Especifica que el dueño y el grupo tendrán permisos de lectura y escritura, y el resto sólo de lectura por default para cualquier archivo nuevo.
