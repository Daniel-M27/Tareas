## *Probabilidad de una moneda*

---

Se pide calcular la probabilidad de que caiga *sol* o *águila* al tirar una moneda N veces y graficar el resultado. Para calcular la probabilidad de que algo ocurra, nos fijamos en el número de casos ṕsibles y el número de casos favorables (en cuántos casos ocurre lo que queríamos). Entonces la probabilidad está dada por el cociente:

$$
P = \frac{\text{Casos Favorables}}{\text{Casos Totales}}
$$

Para las tres simulaciones se utilizará la función `randint(n, m)` de la biblioteca `random`, que toma un número entero $x$ al azar tal que $n \leq x \leq m$. <br>
También, para realizar las simulaciones de manera más rápida, se pedirá al usuario que especifique un *límite superior* y un  *paso*, por ejemplo 1000 y 10 en el programa de la moneda, y lo que hará el programa será tirar la moneda 10 veces, luego 20, 30, ... , etc. Hasta llegar al valor límite especificado. Tras cada conjunto de tiros se toman las probabilidades y se almacenan para graficar los puntos.

En este caso el evento es sencillo, pues solo hay dos casos. Entonces la probabilidad de sacar, tanto águila como sol, al tirar una moneda es $\frac{1}{2}$; Que es el valor al que deben converger los puntos de la simulación. Para simular esto, podemos tomar un valor aleatorio del conjunto $\{0, 1\}$, donde 0 se identifica con sol y 1 con águila.


```python
import random as rd
import matplotlib.pyplot as plt

lim = int(input("límite de tiros: "))
step = int(input("Paso: "))

if lim <= 0 or step < 1:
    raise ValueError("Ingrese un número mayor a 0")
# Add step error

throws = []
prob_ag = []
prob_sol = []

for throw_num in range(step, lim + step, step):
    ag = []
    sol = []
    for i in range(throw_num):
        result = rd.randint(0, 1)
        if result == 0:
            sol.append(result)
        else:
            ag.append(result)
    # Tira una moneda el número de veces especificadas por el usuario.
    # 0 es sol y 1 es águila. Este ciclo genera los datos para un solo
    # punto de la gráfica.

    throws.append(throw_num)
    prob_ag.append(len(ag) / throw_num)
    prob_sol.append(len(sol) / throw_num)
    # Guardamos las probabilidades junto con su respectivo número de tiros.

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(throws, prob_ag, '.b')
ax1.axhline(y = 1/2, color='#444444', linestyle='--',
            label='Valor de Convergencia (1/2)')

ax1.set_ylabel("Probabilidad")
ax1.set_title("Probabilidad de Águila")

ax2.plot(throws, prob_sol, '.r')
ax2.axhline(y = 1/2, color='#444444', linestyle='--',
            label='Valor de Convergencia (1/2)')

ax2.set_xlabel("Número de tiros")
ax2.set_ylabel("Probabilidad")
ax2.set_title("Probabilidad de Sol")

plt.tight_layout()

plt.show()
```

## *Probabilidad de un dado*

---

Este programa es exactamente igual, solo que ahora tomamos un entero entre 1 y 6 con `randint(1,6)`, y la probilidad de sacar 3 es $\frac{1}{6}$, pues hay 6 escenarios posibles.

```python
lim = int(input("límite de tiros: "))
step = int(input("Paso: "))

if lim <= 0 or step < 1:
    raise ValueError("El límite debe ser mayor a 0, y el paso mayor o igual a 1")
# Add step error

throws = []
prob = []

for throw_num in range(step, lim + step, step):
    three = []
    for i in range(throw_num):
        result = rd.randint(1, 6)
        if result == 3:
            three.append(result)
    # Tira un dado el número de veces especificadas en el rango.
    # Si cae 3 guardamos el resultado.
    throws.append(throw_num)
    prob.append(len(three) / throw_num)
# Guardamos la probabilidad y el número de tiros

plt.plot(throws, prob, '.', color='#222288')
plt.axhline(y = 1/6, color='#444444', linestyle='--',
            label='Valor de Convergencia (1/6)')

plt.grid(b=True)
plt.xlim(0, lim)
plt.ylim(0, 1)

plt.xlabel("Tiros")
plt.ylabel("Probabilidad")
plt.title("Probabilidad de que caiga 3")
plt.legend()

plt.tight_layout()

plt.show()
```

## *Suma de dos dados*

La estructura del programa es igual a los anteriores, solo que calcular la probabilidad exacta de que al tirar dos dados sumen 8 es un poco más tardado. Los casos posibles son $6^2 = 36$, y para ver en cuántos de ellos se obtiene una suma de 8 podemos usar Python:

```python
casos = []
for i in range(1,7):
    for j in range(1,7):
        cases.append((i,j))
        
print(cases)
```

```python
favorables = []
for i in range(len(cases)):
    if sum(cases[i]) == 8:
        favorables.append(cases[i])
        
print(favorables)
print(len(favorables))
```

Como hay 5 casos en los que se obtiene 8, y 36 casos totales, la probabilidad es de $\frac{5}{36}$. Y, por lo tanto, la probabilidad de que no sumen 8 es $\frac{31}{36}$. Una vez obtenidos los valores de convergencia podemos graficar los puntos:

```python
lim = int(input("límite de tiros: "))
step = int(input("Paso: "))

if lim <= 0 or step < 1:
    raise ValueError("El límite debe ser mayor a 0, y el paso mayor o igual a 1")

throws = []
prob = []
not_prob = []

for throw_num in range(step, lim + step, step):
    sum = []
    for i in range(throw_num):
        d1 = rd.randint(1, 6)
        d2 = rd.randint(1, 6)
        if d1 + d2 == 8:
            sum.append(d1)
    # Tira dos dados el número de veces especificadas en el rango.
    # Si suman 8, guardamos d1. El valor de d1 no importa, simplemente la
    # estamos usando para contar cuántas veces se da la suma que queremos.

    throws.append(throw_num)
    prob.append(len(sum) / throw_num)
    not_prob.append(1 - len(sum) / throw_num)

    # Guardamos el número de tiros y las probabilidades de los eventos.

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(throws, prob, '.', color='#222288')
ax1.axhline(y = 5/36, color='#444444', linestyle='--',
            label='Valor de Convergencia (5/36)')

ax1.grid(b=True)
ax1.set_xlim(0, lim)
ax1.set_ylim(0, 1)

ax1.set_ylabel("Probabilidad")
ax1.set_title("Probabilidad de que sumen 8")
ax1.legend()

ax2.plot(throws, not_prob, '.r')
ax2.axhline(y = 31/36, color='#444444', linestyle='--',
            label='Valor de Convergencia (31/36)')

ax2.grid(b=True)
ax2.set_xlim(0, lim)
ax2.set_ylim(0, 1)

ax2.set_xlabel("Tiros")
ax2.set_ylabel("Probabilidad")
ax2.set_title("Probabilidad de que NO sumen 8")
ax2.legend()

plt.tight_layout()

plt.show()
```
